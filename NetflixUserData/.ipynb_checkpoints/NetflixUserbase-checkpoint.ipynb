```python
import numpy as np 
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:/PyCharmProject/NetflixUserData/NetflixUserbase.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['Join Date'] = df['Join Date'].astype('datetime64[ns]')
```

```python
df.head()
```

```python
df['Last Payment Date'] = df['Last Payment Date'].astype('datetime64[ns]')
```

```python
df.head()
```

```python
df_male=df[df['Gender']=='Male']
```

```python
df_female=df[df['Gender']=='Female']
```

```python
df_male.head()
```

```python
df_female.head()
```

```python
gender_ratio=df['Gender'].value_counts()
plt.figure(figsize=(7,7))
gender_ratio.plot(kind='pie',autopct='%1.1f%%', textprops={'fontsize':14})
plt.title('Gender Ratio')
plt.show()                                              
```

```python
male_sub_type=df_male['Subscription Type'].value_counts()
plt.figure(0,figsize=(3,3))
male_sub_type.plot(kind='pie',autopct='%1.1f%%', textprops={'fontsize':14})
plt.title('Male Subscription Type')    
plt.ylabel('')

female_sub_type=df_female['Subscription Type'].value_counts()
plt.figure(1,figsize=(3,3))
female_sub_type.plot(kind='pie',autopct='%1.1f%%', textprops={'fontsize':14})
plt.title('Female Subscription Type')
plt.ylabel('')
plt.show() 
```

```python
all_sub_type=df['Subscription Type'].value_counts()

fig,(ax1,ax2,ax3) = plt.subplots(1,3, figsize=(7,7))
fig.tight_layout()

labels=df['Subscription Type'].unique()
ax1.pie(male_sub_type, labels=labels,autopct='%1.1f%%')
ax1.set_title('Male sub type')

labels2=df_female['Subscription Type'].unique()
ax2.set_title('Female sub type')
ax2.pie(female_sub_type, labels=labels, autopct='%1.1f%%')

ax3.set_title('All sub type')
ax3.pie(all_sub_type, labels=labels, autopct='%1.1f%%')

plt.show()
```

```python
all_device_type=df['Device'].value_counts()
male_device_type=df_male['Device'].value_counts()
female_device_type=df_female['Device'].value_counts()
```

```python
labels=df['Device'].unique()
fig,(ax1,ax2,ax3) = plt.subplots(1,3, figsize=(10,10))
fig.tight_layout(pad=5.0)

ax1.pie(all_device_type, labels=labels,autopct='%1.1f%%')
ax2.pie(male_device_type, labels=labels,autopct='%1.1f%%')
ax3.pie(female_device_type, labels=labels,autopct='%1.1f%%')

ax1.set_title('All Device Type')
ax2.set_title('Male Device Type')
ax3.set_title('Female Device Type')

plt.show()
```

```python
country_grouped=df['Country'].value_counts()
country_grouped.plot(kind='bar')
plt.title('Data by country')
plt.ylabel('No. users')
plt.xlabel('Country')
plt.show()
```

```python
df_country_gender=df.groupby(by='Country')['Gender'].value_counts(normalize=True)
```

```python
df_country_gender.plot(kind='bar')
```

```python
df_country_male_ratio=df_country_gender.xs("Male", level="Gender")
```

```python
df_country_male_ratio.plot(kind='bar')
plt.title('Netflix userbase male to female ratio by country ')
plt.ylabel('Ratio')
plt.xlabel('Country')
plt.show()
```
